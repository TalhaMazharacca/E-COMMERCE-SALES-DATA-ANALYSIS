```python
import pandas as pd
data=pd.read_excel("ESD.xlsx")
#print(data)
data.loc[(data["Bonus %"]== 0),"GetBonus"]="NO Bonus" # add column 
data.loc[(data["Bonus %"] > 0),"GetBonus"]= "Bonus"
print(data.head(10))
```
