```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

x=[1,2,3,4,5]
y=[30,70,50,60,40]
y1=[40,25,30,35,20]
y2=[45,27,35,43,20]
y3=[45,27,70,50,60]

plt.figure(figsize=(15,5))
plt.plot(x,y,label="male")
plt.plot(x,y1,label="Female")
plt.plot(x,y2,label="Kids")
plt.plot(x,y3,label="Blank")
plt.legend(["M","F","K","B"],bbox_to_anchor=(0.8,1.1),labelspacing=2,loc = 3,ncol=2)# 1 upto 10
plt.show()
    
```
