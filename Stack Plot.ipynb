```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

days = [1,2,3,4,5,6,7]
NOP1 = [5,10,30,20,35,60,80]
NOP2 = [10,5,35,25,30,55,70]
NOP3 = [15,10,36,28,35,60,80]
color=["HotPink","blue","Magenta"]
plt.stackplot(days,NOP1,NOP2,NOP3,baseline="sym",colors=color,labels = ["weak 1","weak 2","weak 3"])
plt.legend()
plt.show()
```
