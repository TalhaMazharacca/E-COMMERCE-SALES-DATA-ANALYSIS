```python
#add size
import matplotlib.pyplot as plt
y=[98,67,96,74,85]
x=["part 1","part 2","part 3","part 4","part 5"]

plt.bar(x , y)# bar chart
plt.title("Popularity of Harry Potter Movies",fontsize =17)
plt.xlabel("harry Porter Parts",fontsize =17)
plt.ylabel("Popularity %",fontsize =17)
```

```python
#add color
import matplotlib.pyplot as plt
y=[98,67,96,74,85]
x=["part 1","part 2","part 3","part 4","part 5"]
color=["yellow","brown","pink","blue","green"]
plt.bar(x , y,color=color,edgecolor="Red")# bar chart
plt.title("Popularity of Harry Potter Movies",fontsize =17)
plt.xlabel("Harry Porter Parts",fontsize =17)
plt.ylabel("Popularity %",fontsize =17)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data1=pd.read_excel("expense3.xlsx")
df= pd.DataFrame(data1)
print(df)
print()
print()
print()
group_by = df.groupby("Payment Mode")["Amount"].sum()
print(group_by)
print()
print()
print()

color=["yellow","brown","green"]
plt.bar(group_by.index,group_by.values,color=color,edgecolor="black")
plt.title("Mode Vs Payment",fontsize =17)
plt.ylabel("Amounts",fontsize =15)
plt.xlabel("Modes",fontsize =15)

plt.show()
```
