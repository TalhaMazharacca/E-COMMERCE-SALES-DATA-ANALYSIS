```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x= np.random.randint(1,10,20)
y= np.random.randint(10,100,20)
color=np.random.randint(10,100,20)

# plt.scatter(x,y, marker="*",cmap="viridis",c=color)
plt.scatter(x,y, marker="*",cmap="hot",c=color)
plt.colorbar()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x= np.random.randint(1,10,20)
y= np.random.randint(10,100,20)
color=np.random.randint(10,100,20)
size=np.random.randint(10,100,20)

# plt.scatter(x,y, marker="*",cmap="viridis",c=color)
plt.scatter(x,y, marker="*",cmap="hot",c=color,s=size)
plt.colorbar()
plt.show()
```
