```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

x=["Day 1","Day 2","Day 3","Day 4","Day 5"]
y=[30,40,50,30,35]

plt.step(x,y, where="pre",marker="o")
plt.show()

plt.step(x,y, where="mid")
plt.show()

plt.step(x,y, where="post")
plt.show()
```
