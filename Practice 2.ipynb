```python
import pandas as pd
data1={"Emp ID":["ED1","ED2","ED3","ED4","ED5","ED6"],
       "Name":["AB1","AB2","AB3","AB4","AB5","AB6"],
       "Age":[23,24,25,26,27,29]}
data2={"Emp ID":["ED1","ED2","ED3","ED4","ED5","ED6"],
       "Salary":[23000,24000,25000,26000,27000,29000]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
print(df1)
print(df2)

```

```python
print(pd.merge(df1,df2,on ="Emp ID"))
```

```python
import pandas as pd
data1={"Emp ID":["ED1","ED2","ED3","ED4","ED5","ED6"],
       "Name":["AB1","AB2","AB3","AB4","AB5","AB6"],
       "Age":[23,24,25,26,27,29]}
data2={"Emp ID":["ED1","ED7","ED3","ED8","ED5","ED6"],
       "Salary":[23000,24000,25000,26000,27000,29000]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
print(df1)
print(df2)
```

```python
print(pd.merge(df1,df2,on="Emp ID",how="left"))
print()
print()
print(pd.merge(df2,df1,on="Emp ID",how="left"))
print()
print()
print(pd.merge(df1,df2,on="Emp ID",how="right"))
print()
print()
print(pd.merge(left=df1,right=df2,on="Emp ID",how="left"))
print()
print()
print(pd.merge(right=df1,left=df2,on="Emp ID",how="left"))
```

```python
import pandas as pd

# Assuming d1 is your DataFrame
# Example DataFrame:
data = {
    "keys": ["k1", "k2", "k1", "k2"],
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Houses": [1, 2, 3, 4]
}

d1 = pd.DataFrame(data)

# Perform the pivot operation
pivoted_df = d1.pivot(index="keys", columns="Name", values="Houses")

# Print the pivoted DataFrame
print(pivoted_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

x=["Day 1","Day 2","Day 3","Day 4","Day 5"]
y=[30,40,50,30,35]

plt.step(x,y,where="pre")
plt.show()

plt.step(x,y,where="mid")
plt.show()

plt.step(x,y,where="post")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_excel("expense3.xlsx")
df=pd.DataFrame(data)

gp=df.groupby("Category").agg({"Amount":"sum"})


print(gp)

plt.step(df["Category"].unique(),gp["Amount"])
plt.show()
```
