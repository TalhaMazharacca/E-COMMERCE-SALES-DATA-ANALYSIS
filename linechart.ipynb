```python
import matplotlib.pyplot as plt

x = ["day1","day2"," day3","day4","day5"]
y = [300,400,350,432,250]
plt.plot(x,y,marker=".",ls=":",color="red")# shift + tap for more funtion
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["day1","day2"," day3","day4","day5"]
y = [300,400,350,432,250]
y1 = [350,370,360,479,269]
plt.plot(x,y,marker=".",ls=":",color="red",label="week1")# shift + tap for more funtion
plt.plot(x,y1,marker=".",ls="-",color="green",label="week2",alpha = 0.25)#alpha to make lighter or transparancy
plt.legend()# shift + tap for more funtion
plt.show()
```
