```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

x=[10,0,40,50,60,70,20,10,30,40,50,60,70]
plt.stem(x,linefmt="--",markerfmt="^",bottom = 1)
plt.show()
```
