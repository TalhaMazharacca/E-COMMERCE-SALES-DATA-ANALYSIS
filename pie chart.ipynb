```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_excel("expense3.xlsx")
df=pd.DataFrame(data)
print(df)
grouped_by=df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)

plt.pie(grouped_by.values,labels = grouped_by.index,autopct="%.2f")
plt.title("Paymode %")
plt.show()
```
