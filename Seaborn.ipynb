```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df=sns.load_dataset("exercise")
print(df)
sns.set_style(style="darkgrid")
sns.barplot(data=df,x="time",y="pulse")
plt.show()

sns.set_style(style="whitegrid")
sns.barplot(data=df,x="time",y="pulse")
plt.show()

sns.set_style(style="ticks")#small line on each axis
sns.barplot(data=df,x="time",y="pulse")
plt.show()
```
